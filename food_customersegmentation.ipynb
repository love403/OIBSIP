## K CLUSTER

```python
from sklearn.cluster import KMeans
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
food['In_relationship'] = food.apply(get_relationship, axis=1)
food.head()   
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship']
food_scaled = food.copy()
food_scaled[cols_for_clustering] = scaler.fit_transform(food[cols_for_clustering])
food_scaled[cols_for_clustering].describe()
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(food_scaled[cols_for_clustering])
food_scaled['pc1'] = pca_res[:,0]
food_scaled['pc2'] = pca_res[:,1]
```

```python
X = food_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 200, 200 / 200, 130 / 200))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

## CREATING Visualisation of clusters using scattered and bar chart 

```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(food_scaled[cols_for_clustering])
food_scaled['Cluster'] = model.predict(food_scaled[cols_for_clustering])
```

```python
sns.scatterplot(x='pc1', y='pc2', data=food_scaled, hue ='Cluster', palette='cool')
plt.title('Visualisation by clusters')
```

```python
food['Cluster'] = food_scaled.Cluster
food.groupby('Cluster')[cols_for_clustering].mean()
```
